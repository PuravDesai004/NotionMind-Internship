# Defining the role


````python
from anthropic import Anthropic
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv("ANTHROPIC_API_KEY")
client = Anthropic(api_key=api_key)
model = "claude-haiku-4-5-20251001"

patient_data = """
P001: Female,45,BMI31.2,BP145/92,Glucose128,Chol240,NonSmoker,Exercise1
P002: Male,29,BMI24.1,BP118/76,Glucose92,Chol180,Smoker,Exercise2
P003: Female,62,BMI28.5,BP158/96,Glucose145,Chol260,NonSmoker,Exercise0
"""

prompt = f"""Analyze this patient data and identify health risks:
{patient_data}"""

system = "Analyze the patient dataset and identify common health patterns, key risk factors, and regional health concerns. Summarize the main findings and areas needing attention.."

messages = [{"role": "user", "content": prompt}]

response = client.messages.create(
    model=model,
    max_tokens=1500,
    messages=messages,
    system=system  
)

print(response.content[0].text)
````

# Healthcare Analysis Prompt Comparison

This notebook compares multiple system prompt versions for analyzing a small patient dataset with BMI, blood pressure, glucose, cholesterol, smoking status, and exercise frequency.

## Dataset

The analysis uses three patient records:

- P001: Female, 45, BMI 31.2, BP 145/92, Glucose 128, Cholesterol 240, Non-smoker, Exercise 1
- P002: Male, 29, BMI 24.1, BP 118/76, Glucose 92, Cholesterol 180, Smoker, Exercise 2
- P003: Female, 62, BMI 28.5, BP 158/96, Glucose 145, Cholesterol 260, Non-smoker, Exercise 0

## Prompt Version Comparison

| Version | Prompt Focus | Result Quality | Key Notes |
|---|---|---|---|
| V1 | Identify health risks in the data | Good | Strong patient-wise risk breakdown with clear tables and priority recommendations, but less focused on overall dataset patterns. |
| V2 | Explain key health risks in simple words | Good for beginners | Very readable and easy to understand, but less detailed and less structured for a formal healthcare analysis. |
| V3 | Patient-wise risks, health patterns, and hospital visit yes/no | Useful but too clinical | Includes clear action decisions, but the hospital visit labels may be too strong without more clinical context. |
| V4 | Summarize patterns, common risk factors, high-priority patients, and key concerns | Best | Best balance of clarity, structure, patient prioritization, common patterns, and brief summary. It directly matches the analysis goal. |
| V5 | Common patterns, risk factors, regional health concerns, areas needing attention | Mixed | Has a strong table and recommendations, but includes an irrelevant "regional health concerns" requirement and a factual issue about hypertension across all patients. |

## Best Performing Prompt

**V4 gave the best result.**

V4 worked best because it asked for:

- Main health patterns
- Common risk factors
- Patients needing the most attention
- Key health concerns
- A brief final summary

This produced the most complete and balanced output. It identified P003 as the highest-priority patient, P001 as moderate priority, and P002 as lower priority with smoking as the main concern.

## Final Recommendation

Use V4 as the preferred system prompt style for this healthcare analysis task:

```text
Analyze the patient dataset and summarize the main health patterns you observe. Identify any common risk factors, highlight the patients who may need the most attention, and explain the key health concerns shown in the data. Provide a brief summary of your findings.
```

This prompt is specific enough to guide the model while still leaving room for a useful, structured analysis.
