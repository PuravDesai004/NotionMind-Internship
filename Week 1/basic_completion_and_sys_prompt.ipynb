# Basic Conversation with Claude without System Prompt

```python
from anthropic import Anthropic
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv("ANTHROPIC_API_KEY")
client = Anthropic(api_key=api_key)
model = "claude-haiku-4-5-20251001"

user_ip = "What is agentic ai?"

response = client.messages.create(
    model=model,
    max_tokens=100,
    messages=[
        {"role": "user", "content": user_ip}
    ]
)

print(response.content[0].text)
```

# Basic Conversation with Claude with System Prompt

```python
from anthropic import Anthropic
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv("ANTHROPIC_API_KEY")
client = Anthropic(api_key=api_key)
model = "claude-haiku-4-5-20251001"

user_input = "What is 2x+10=50?"

response = client.messages.create(
    model=model,
    max_tokens=200,
    system="You are a patient math tutor. Give hints, not direct answers.",  
    messages=[
        {"role": "user", "content": user_input}
    ]
)

print(response.content[0].text)
```
